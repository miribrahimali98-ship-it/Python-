#  Enterprise Sales Analytics: Revenue, Profit & Discount Insights

##  Project Overview
This project analyzes enterprise-level sales data to uncover key insights related to **revenue trends**, **profit performance**, **regional sales distribution**, and the **impact of discounting on customer orders**.  
The goal is to support **data-driven business decisions** using Python-based data analysis and visualization.

---

##  Business Objectives
- Analyze **Month-over-Month (MoM)** and **Year-over-Year (YoY)** revenue trends
- Identify **top-performing regions** based on revenue and profit
- Understand how **discount levels influence order volume**
- Highlight patterns that can guide **pricing and sales strategy**

## Key Analyses Performed
1. **Revenue Month-over-Month Analysis**  
   Identified seasonal trends and revenue fluctuations across months.

2. **Revenue Year-over-Year Analysis**  
   Compared annual revenue growth to evaluate business performance over time.

3. **Revenue & Profit by Region**  
   Assessed regional contribution to total revenue and profitability.

4. **Impact of Discount on Orders**  
   Analyzed how varying discount levels affect customer order volume.

---

## Key Insights
- Revenue shows noticeable variation across months, indicating possible seasonality.
- Certain regions consistently outperform others in both revenue and profit.
- Higher discounts do not always result in proportionally higher order volumes.
- An optimal discount range exists where orders increase without significantly impacting profit.

---

## Conclusion
This analysis demonstrates how sales data can be transformed into actionable insights using Python.  
The findings can help stakeholders optimize **pricing strategies**, **regional focus**, and **sales planning**.

---

#Revenue Month over Month

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/Desktop/Analytics/Enterprise_Sales_Analytics/revenue_mom.csv")
n = data[["month", "revenue"]]
df = pd.DataFrame(n)
plt.bar(df["month"],df["revenue"],width = 0.5)
plt.xticks(rotation = 90)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

#Revenue Year over Year

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/Desktop/Analytics/Enterprise_Sales_Analytics/revenue_YoY.csv")
n = data[["year", "revenue"]]
df = pd.DataFrame(n)
plt.bar(df["year"],df["revenue"],width = 0.5)
plt.xticks(rotation = 90)
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

#Revenue & profit by region

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/Desktop/Analytics/Enterprise_Sales_Analytics/Revenue & profit by region.csv")
n = data[["region", "revenue","profit"]]
df = pd.DataFrame(n)
plt.bar(df["region"],df["revenue"],width = 0.5,color = "r")
plt.bar(df["region"],df["profit"],width = 0.3,color = "m")
plt.xticks(rotation = 90)
plt.xlabel("region")
plt.ylabel("Revenue")
plt.show()
```

#Impact of Discount on Profit

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/Desktop/Analytics/Enterprise_Sales_Analytics/Discount impact on profit.csv")
n = data[["discount", "orders"]]
df = pd.DataFrame(n)
plt.figure(figsize=(10,5))
plt.plot(df["discount"], df["orders"], marker='o')
plt.title("Number of Orders by Discount")
plt.xlabel("Discount")
plt.ylabel("Orders")
plt.grid(True)
plt.show()
```

#Average Profit By Discount 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/Desktop/Analytics/Enterprise_Sales_Analytics/Discount impact on profit.csv")
n = data[["discount", "avg_profit"]]
df = pd.DataFrame(n)
plt.figure(figsize=(10,5))
plt.plot(df["discount"], df["avg_profit"], marker='o')
plt.title("Average Profit by Discount")
plt.xlabel("Discount")
plt.ylabel("Average Profit")
plt.grid(True)
plt.show()
```

#Impact of Discount on Orders and Profit

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ASUS/Desktop/Analytics/Enterprise_Sales_Analytics/Discount impact on profit.csv")
n = data[["discount","orders","avg_profit"]]
df = pd.DataFrame(n)
fig, ax1 = plt.subplots(figsize=(10,5))

ax1.plot(df["discount"], df["orders"], marker='o')
ax1.set_xlabel("Discount")
ax1.set_ylabel("Orders")

ax2 = ax1.twinx()
ax2.plot(df["discount"], df["avg_profit"], marker='s')
ax2.set_ylabel("Average Profit")

plt.title("Impact of Discount on Orders and Profit")
plt.grid(True)
plt.show()
```
